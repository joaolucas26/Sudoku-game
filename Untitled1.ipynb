```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xtrain = 10*np.random.normal(0,1, 70)
ytrain = 10*(-xtrain**2) + np.random.normal(-100,100,70)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(xtrain, ytrain, s=15)
plt.xlabel('feature', fontsize=16)
plt.ylabel('response', fontsize=16)
plt.show()
```
